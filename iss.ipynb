```python
import matplotlib.pyplot as plt
import requests
import time

latitude = []
longitude = []
timestamp = []

for x in range(0,100):
    #request data
    data = []
    r = requests.get("http://api.open-notify.org/iss-now.json")
    json_data = r.json()
    
    #append data
    data.append(json_data["iss_position"]["latitude"])
    data.append(json_data["iss_position"]["longitude"])
    data.append(json_data["timestamp"])
    print(data)
    
    #path of iss
    latitude.append(json_data["iss_position"]["latitude"])
    longitude.append(json_data["iss_position"]["longitude"])
    timestamp.append(json_data["timestamp"])
    
    #ping every 10 secs
    time.sleep(10)
```

```python
plt.plot(longitude, latitude, '-o')
plt.title('ISS Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.xticks(range(0, 101, 19))
plt.yticks(range(0, 101, 19))

plt.xticks(fontsize=7)
plt.yticks(fontsize=7)

plt.show()
```
